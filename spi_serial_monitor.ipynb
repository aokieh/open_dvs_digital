```python
import serial
import time
```

```python
# Set the serial port to the same one that your Arduino is connected to
serial_port = 'COM4'  # Replace with your actual port, e.g., 'COM3' on Windows
baud_rate = 9600  # Must match the baud rate of your Arduino

# Open the serial port
ser = serial.Serial(serial_port, baud_rate)

# Open the file to write the data
output_file = "output_data.txt"
```

```python
# # Function to capture and clean data
# def capture_data():
#     with open(output_file, "w") as file:
#         print("Starting to capture data...")

#         data_buffer = []

#         try:
#             while True:
#                 line = ser.readline().decode('utf-8').strip()  # Read one line of data from serial port
#                 if line:
#                     # Example: "17:49:03.591 -> Addr[0]: D0"
#                     # We want to capture only "Addr[0]: D0"
                    
#                     # Find the part that contains "Addr[x]: data"
#                     if "Addr" in line:
#                         # Split the line to extract just the address and data
#                         addr_data = line.split("->")[-1].strip()  # Remove timestamp and '->'
                        
#                         # Write the filtered data to the file
#                         file.write(addr_data + "\n")
#                         print(addr_data)  # Print to console

#         except KeyboardInterrupt:
#             print("Data capture stopped.")

#     print(f"Data saved to {output_file}")
#     ser.close()

# # Run the function
# # capture_data

# Function to capture and clean data
def capture_data():
    with open(output_file, "w") as file:
        print("Starting to capture data...")

        try:
            while True:
                # Read one line of data from serial port and decode it
                line = ser.readline().decode('utf-8').strip()
                
                if line:
                    # 1. Look for the start of the final read-out line: "Addr[#"
                    if line.startswith("Addr["):
                        # 2. Split the line to isolate the address and data part.
                        # This handles cases where a timestamp/prefix might exist (like from an external serial monitor)
                        # but based on your dump, it mainly cleans up the line itself.
                        # It also handles the single line "Addr[0]: D0" format
                        if "]:" in line:
                            # This is the line we want: e.g., "Addr[0]: D0"
                            
                            # Clean up any potential leading junk (e.g., timestamps)
                            if "->" in line:
                                addr_data = line.split("->")[-1].strip()
                            else:
                                addr_data = line.strip()
                            
                            # Write the cleaned data to the file
                            file.write(addr_data + "\n")
                            print(addr_data)  # Print to console
                            
        except KeyboardInterrupt:
            print("Data capture stopped.")

    print(f"Data saved to {output_file}")
    ser.close()
```

```python
# Run the function
capture_data()
```
